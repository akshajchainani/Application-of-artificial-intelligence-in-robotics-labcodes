```python
df= pd.read_csv("/content/drive/MyDrive/Datasets AIR/User_Data.csv")
df
```

```python
#select age and estimated salary as input(dependent variable) predict if a person has made a purchase or not
x= df.iloc[:,[2,3]]
x
```

```python
#select "purchased" as the target variable
y= df.iloc[:,[-1]]
y
```

```python
#splitting the dataset into the training and test set
from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test= train_test_split(x,y,test_size= 0.3,random_state=40)
```

```python
#feature scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_x_train = scaler.fit_transform(x_train)
scaled_x_test= scaler.fit_transform(x_test)
```

```python
#use naive bayes algorithm for classification
from sklearn.naive_bayes import GaussianNB
classifier= GaussianNB()
classifier.fit(scaled_x_train, y_train)
```

```python
#predicitng the test results
y_pred = classifier.predict(scaled_x_test)
```

```python
#making confusion matrix
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test, y_pred)
cm
```
