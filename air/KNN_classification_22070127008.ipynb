```python
df.drop(columns= ['Unnamed: 32','id'],axis=1 ,inplace=True)
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
df.diagnosis.unique()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['diagnosis'] = label_encoder.fit_transform(df['diagnosis'])
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

```

```python
df.columns
```

```python
X = (df.drop('diagnosis',axis =1))
Y = df['diagnosis']
Y
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2,random_state = 101)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=11)
model
```

```python
sacled_x_train = scaler.fit_transform(x_train)
scaled_x_test = scaler.fit_transform(x_test)
```

```python
model.fit(sacled_x_train,y_train)
```

```python
prediction = model.predict(scaled_x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
accuracy_score(prediction,y_test)
```

```python
classification_report(y_test,prediction)
```

```python
confusion_matrix(y_test,prediction)
```

```python
test_error_rate = []
for k in range(1,30):
  knn_model = KNeighborsClassifier(n_neighbors=k)
  knn_model.fit(sacled_x_train,y_train)

  pred = knn_model.predict(scaled_x_test)

  test_error = 1-accuracy_score(y_test,pred)
  test_error_rate.append(test_error)

test_error_rate
```

```python
plt.figure(figsize=(8,5),dpi = 200)
plt.plot(range(1,30),test_error_rate,label='Test Error')
plt.legend()
plt.xlabel('K value')
plt.ylabel('Error rate')
```
