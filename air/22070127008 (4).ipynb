```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set seed for reproducibility
np.random.seed(42)

# Generate data
data = {
    'IQ': np.random.normal(loc=100, scale=15, size=1000),  # Generating IQ scores with mean 100 and standard deviation 15
    'CGPA': np.random.uniform(2.0, 4.0, 1000),  # Generating CGPA between 2.0 and 4.0
    'Placement': np.random.choice(['Placed', 'Not Placed'], size=1000)  # Random placement status
}

# Create DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
df
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['Placement'] = label_encoder.fit_transform(df['Placement'])
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(df.drop(columns = ['Placement']))
Y = df.drop(columns = ['IQ','CGPA'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.4,random_state = 101)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier()
model
```

```python
model.fit(x_train,y_train)
```

```python
prediction = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(prediction,y_test)
```

```python
from sklearn.svm import SVC
model = SVC()
model
```

```python
model.fit(x_train,y_train)
```

```python
prediction2 = model.predict(x_test)
```

```python
r2_score(prediction2,y_test)
```
