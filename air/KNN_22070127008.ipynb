# KNN Imputer for qualitative or categorical data

## Import or create required dataset

```python
dz={'Name': ['Root','Stark','John','Marsh','Sam'],'Age': [31, 34, 33, 35,32],'Department': ['Navy','Army',np.nan,'Navy',np.nan],'Color_Code': ['white', 'Green', 'White',np.nan,'White']}
```

```python
dz=pd.DataFrame(dz)
dz
```

```python
dz.to_csv("/content/drive/MyDrive/AIR_LAB_DATA/dz")
```

## Convert the string to float

```python
dz.columns
```

### Map with lables

```python
dz
dz['Department_map']=dz.Department.map({'Navy':0,'Army':1})
dz['Color_Code_map']=dz.Color_Code.map({'White':0,'Green':1})
```

```python
dz
```

```python
dz1=dz[['Age','Department_map']]
dz1
```

### Apply KNN imputer

```python
imputer=KNNImputer(n_neighbors=1)
dz2=imputer.fit_transform(dz1)
dz2
```
