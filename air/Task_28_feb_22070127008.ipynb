```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Dataster_AAI/KNN_Task_Dataset.csv')
df
```

```python
df.head()
```

```python
df.drop(columns = 'Unnamed: 0',inplace=True)
```

```python
df.drop(columns = 'Unnamed: 1',inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['Diagnosis'] = label_encoder.fit_transform(df['Diagnosis'])
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(columns = 'Diagnosis')
y = df['Diagnosis']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaled_x_train = scaler.fit_transform(x_train)
scaler_x_test = scaler.fit_transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(scaled_x_train,y_train)
```

```python
prediction = knn.predict(scaler_x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
accuracy_score(y_test,prediction)
```

```python
test_error_rate = []
for k in range(1,30):
  knn_model = KNeighborsClassifier(n_neighbors=k)
  knn_model.fit(scaled_x_train,y_train)

  pred = knn_model.predict(scaler_x_test)

  test_error = 1-accuracy_score(y_test,pred)
  test_error_rate.append(test_error)

test_error_rate
```

```python
plt.figure(figsize=(8,5),dpi = 200)
plt.plot(range(1,30),test_error_rate,label='Test Error')
plt.legend()
plt.xlabel('K value')
plt.ylabel('Error rate')
```

```python
model = KNeighborsClassifier(n_neighbors=1)
model.fit(scaled_x_train,y_train)
pred_1 = model.predict(scaler_x_test)
```

```python
accuracy_score(y_test,pred_1)
```
